## Descripción

En el mundo del desarrollo independiente de videojuegos, no solo importa qué haces, sino cuándo y dónde lo lanzas. Un buen juego puede pasar desapercibido si se lanza en el momento o plataforma equivocados.

Este análisis exploratorio de datos (EDA) se busca identificar patrones que ayuden a responder una pregunta clave para cualquier estudio indie:

¿Cuál es la mejor temporada, región y plataforma para lanzar un videojuego indie?



### HIPÓTESIS

Los videojuegos indie tienen un mejor desempeño inicial cuando se lanzan en verano.

### HIPÓTESIS SECUNDARIA

Tienen un mejor desempeño especialmente si:

- Son juegos de aventura (en Europa) 
- Son RPGs (en Asia)


Esto sugiere que tanto la temporada, como la región de origen y el género del juego influyen en el impacto de un lanzamiento.
